```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from loadTDMS import *
import pylab
from matplotlib.ticker import *
```

```python
last_t = 0

path = r'C:\Users\pmg07\Desktop\2021_1115_切削実験\dyanano\Dynamometer_50.tdms'
# path = r'Z:\100 実験データ（RAW）\20211115---切削実験---AE+切削抵抗\Dynamometer\1-5000\Dynamometer_50.tdms'


d = load_tdms(path, ch_num=3)
print(len(d[0]))
dt = 1/(10**6)*3
N = len(d[0])
t_d = np.linspace(last_t, last_t + N*dt, N)

print(len(t_d))
```

```python
start = 0
end = 9000*500

print(t_d[end])

fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(111)
# p = plt.plot(t_d[start:end], d[0][start:end], linewidth = 3.0)
p = plt.plot(t_d[start:end], d[1][start:end], linewidth = 3.0)
# p = plt.plot(t_d[start:end], d[2][start:end], linewidth = 3.0)


# p = plt.plot(t_d, d[0], linewidth = 3.0)
# p = plt.plot(t_d, d[1], linewidth = 3.0)
# p = plt.plot(t_d, d[2], linewidth = 3.0)


# ax.set_ylim(-600,500)

p = ax.set_xlabel("Time[ms]", fontsize=20)
p = ax.set_ylabel("Amplitude[V]", fontsize=20)

plt.yticks(np.arange(-0.9, 0.601, step=0.3))


xticks, strs = pylab.xticks()
pylab.xticks(xticks, ["%.f" % x for x in 100 * xticks])
# ax.set_xlim(0, 0.11)
ax.set_xlim(0, 2.6)


p = ax.legend(["送り分力","主分力","背分力"], fontsize=20)
p = plt.tick_params(labelsize=16)
# p = plt.hlines([0],0,0.145)

ax = plt.gca()
# fig.savefig("cutting_force_amplitude_50_10ms.png")
# ax.xaxis.set_major_locator(MultipleLocator(1))  # x軸の場合は yaxis -> xaxis
```
