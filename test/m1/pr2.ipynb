```python
import numpy as np
import pandas as pd
import matplotlib
from nptdms import TdmsFile
import itertools
```

```python
# TDMSファイルからAE信号強度を配列にする
with TdmsFile.open("AE_125.tdms") as tdms_file:
    group_name = "g0"
    channel_name = "AE_Signal"  # AE_Signalチャンネルから
    channel = tdms_file[group_name][channel_name]

    all_data = []

    # データが大きいからchunkで分割して、配列に代入
    for chunk in channel.data_chunks():
        channel_chunk_data = chunk[:]
        # print(len(channel_chunk_data))
        if(len(channel_chunk_data) == 0):
            break

        all_data.append(channel_chunk_data)
```

```python
# all_dataの中身が[[chunkの数], [chunkの数], ・・・・・[chunkの数]]になっていたので
# all_dataを一次元化する
value_d = list(itertools.chain.from_iterable(all_data))
```

```python
time_length = 16.668
```

```python
# 時間を表す配列を作る
time_d = []
# 1サンプリングするのにかかる時間
sample_rate = 1/10**6
```

```python
# 時間を表す配列を作るために、16,548,822(flatten_all_dataの長さ)分だけ
# sample_rate × i をする。ただし、16.668(time_length)を超えない時まで
for i in range(len(value_d)):
    if(time_length < sample_rate*(i-1)):
        break
    time_d.append(sample_rate*(i-1))
```

```python
dict1 = dict(time=time_d[:30000000], value=value_d[:30000000])
dict2 = dict(time=time_d[30000000:60000000], value=value_d[30000000:60000000])
dict3 = dict(time=time_d[60000000:90000000], value=value_d[60000000:90000000])
dict4 = dict(time=time_d[90000000:120000000], value=value_d[90000000:120000000])
dict5 = dict(time=time_d[120000000:150000000], value=value_d[120000000:150000000])
dict6 = dict(time=time_d[150000000:], value=value_d[120000000:])
```

```python
df1 = pd.DataFrame(data=dict1)
df2 = pd.DataFrame(data=dict2)
df3 = pd.DataFrame(data=dict3)
df4 = pd.DataFrame(data=dict4)
df5 = pd.DataFrame(data=dict5)
df6 = pd.DataFrame(data=dict6)
```

```python
df1.to_csv('prrr2.csv')
df2.to_csv('prrr2.csv', mode='a', header=False)
df3.to_csv('prrr2.csv', mode='a', header=False)
df4.to_csv('prrr2.csv', mode='a', header=False)
df5.to_csv('prrr2.csv', mode='a', header=False)
df6.to_csv('prrr2.csv', mode='a', header=False)
```
